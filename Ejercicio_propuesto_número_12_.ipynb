```python
# Paso 1: Ingresar total horas trabajadas
horas = 48

# Paso 2: Ingresar valor por hora
valor_hora = 5000

# Paso 3: Ingresar porcentaje de retención
porcentaje_retencion = 12.5

# Paso 4: Calcular Salario Bruto
salario_bruto = horas * valor_hora
print("Salario Bruto:", salario_bruto)

# Paso 6: Convertir porcentaje a decimal
retencion_decimal = porcentaje_retencion / 100
print("Retención en decimal:", retencion_decimal)

# Paso 8: Calcular valor real de retención
valor_retencion = salario_bruto * retencion_decimal
print("Retención en la fuente:", valor_retencion)

# Paso 10: Calcular Salario Neto
salario_neto = salario_bruto - valor_retencion
print("Salario Neto:", salario_neto)
```
